```python
impact_of_stress_on_productivity_over_gens=df.groupby(['Productivity_Change','Generation']).size().reset_index(name="numof employyes")
```

```python
impact_of_stress_on_productivity_over_gens
```

```python
impact_of_stress_on_productivity_over_gens = (
    df.groupby(['Productivity_Change', 'Generation'])
    .size()
    .reset_index(name="numof_employees")
)

impact_of_stress_on_productivity_over_gens['percentage'] = (
    impact_of_stress_on_productivity_over_gens
    .groupby('Generation')['numof_employees']
    .apply(lambda x: x / x.sum() * 100)
    .values
)

# Ù†Ø®Ù„ÙŠÙ‡Ø§ Ø¨Ø±Ù‚Ù…ÙŠÙ† Ø¹Ø´Ø±ÙŠÙŠÙ†
impact_of_stress_on_productivity_over_gens['percentage'] = (
    impact_of_stress_on_productivity_over_gens['percentage'].round(2)
)

impact_of_stress_on_productivity_over_gens
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Ù‡Ù†ÙØªØ±Ø¶ Ø¥Ù† Ø¹Ù†Ø¯Ùƒ Ø§Ù„Ø¯Ø§ØªØ§ Ø§Ù„Ù„ÙŠ Ø­Ø³Ø¨Ù†Ø§Ù‡Ø§ Ù‚Ø¨Ù„ ÙƒØ¯Ù‡
# impact_of_stress_on_productivity_over_gens

plt.figure(figsize=(10,6))

# Ù†Ø±Ø³Ù… Ø§Ù„Ø¨Ø§Ø±Ø§Øª
ax = sns.barplot(
    data=impact_of_stress_on_productivity_over_gens,
    x='Generation',
    y='percentage',
    hue='Productivity_Change',
    palette='Set2'
)

# Ù†Ø¶ÙŠÙ Ø§Ù„Ø£Ø±Ù‚Ø§Ù… ÙÙˆÙ‚ Ø§Ù„Ø£Ø¹Ù…Ø¯Ø© (Ø²ÙŠ Ø§Ù„Ù„ÙŠ ÙÙŠ Ø§Ù„ØµÙˆØ±Ø©)
for container in ax.containers:
    ax.bar_label(container, fmt='%.2f%%', fontsize=9, padding=3)

# Ù†Ø¶ÙŠÙ Ø§Ù„Ø¹Ù†ÙˆØ§Ù† ÙˆØ§Ù„Ù…Ø­Ø§ÙˆØ±
plt.title('IMPACT OF STRESS ON PRODUCTIVITY OVER GENS', fontsize=14, weight='bold')
plt.xlabel('Productivity over Gens', fontsize=12)
plt.ylabel('Percent of employees', fontsize=12)
plt.ylim(0, 70)  # Ù„Ùˆ Ø¹Ø§ÙŠØ² Ù†ÙØ³ Ù…Ù‚ÙŠØ§Ø³ Ø§Ù„ØµÙˆØ±Ø©
plt.legend(title='Productivity Change')

plt.tight_layout()
plt.show()
```

```python
score_cols = ['Wellbeing Score', 'compny support index', 'wmental', 'wsleep', 'waccess', 'wstress']
```

```python
sns.set_theme(style="whitegrid")
plt.figure(figsize=(10, 8))
corr_matrix = df[score_cols].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix: Drivers of Employee Wellbeing\n(Impact of Sleep, Mental Health, and Support)', fontsize=14, weight='bold')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
job_wellbeing = df.groupby('Job_Role')['Wellbeing Score'].mean().sort_values(ascending=False).reset_index()
#Ø­Ø³Ø§Ø¨ Ø§Ù„Ù…ØªÙˆØ³Ø· Ù„ÙƒÙ„ ÙˆØ¸ÙŠÙØ© ÙˆØªØ±ØªÙŠØ¨Ù‡Ø§ Ù…Ù† Ø§Ù„Ø£Ù‚Ù„ Ù„Ù„Ø£Ø¹Ù„Ù‰
```

```python
sns.barplot(
    data=job_wellbeing, 
    x='Job_Role', 
    y='Wellbeing Score', 
    palette='viridis', 
    hue='Job_Role',    
    legend=False       
)
plt.title('Average Wellbeing Score by Job Role', fontsize=14, weight='bold')
plt.xlabel('Job Role', fontsize=12)
plt.ylabel('Average Wellbeing Score (Higher is Better)', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
support_satisfaction = df.groupby('Satisfaction_with_Remote_Work')['compny support index'].mean().reset_index()
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(
    data=support_satisfaction, 
    x='Satisfaction_with_Remote_Work', 
    y='compny support index', 
    palette='pastel', 
    hue='Satisfaction_with_Remote_Work', 
    legend=False                          
)

plt.title('Link Between Company Support & Remote Work Satisfaction', fontsize=14, weight='bold')
plt.xlabel('Satisfaction Level', fontsize=12)
plt.ylabel('Avg Company Support Index', fontsize=12)
plt.tight_layout()
plt.show()
```
